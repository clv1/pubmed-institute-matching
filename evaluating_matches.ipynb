# Setup

### Imports

```python
import pandas as pd
```

### Load data

```python
df = pd.read_csv("./output_data/df.csv")
df.head()
```

# Evaluation

### What percentage of records have matched? Is this number what you expected?


```python
proper_institutions = df['proper_institution']

invalid_count = proper_institutions.isin(['', 'nan', '(nan,)', 'Institution name not found.']).sum()

percentage = (invalid_count / len(proper_institutions)) * 100
print(percentage)
```

### Pick a sample of 10-20 records that have a match and manually inspect them. Any surprises?

```python
proper_institutions = df['proper_institution']

valid_entries = df[(proper_institutions != '') & (~proper_institutions.isin(['nan', '(nan,)', 'Institution name not found.']))]
sample = valid_entries.sample(20)

print(sample)
```

### Pick a sample of 10-20 records that have not matched and manually inspect them (having the `institutes.csv` open on a side window might help). Any surprises?

```python
proper_institutions = df['proper_institution']

invalid_entries = df[(proper_institutions == '') | (proper_institutions.isin(['nan', '(nan,)', 'Institution name not found.']))]
sample = invalid_entries.sample(20)

print(sample)
```
